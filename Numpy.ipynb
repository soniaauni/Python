```python
import time
numbers = list(range(1, 10000000))   # 1 2 3 4 5 6 7 8 ..........................99999
total = 0
start_time = time.time() # record start time
for n in numbers:
  total += n
end_time = time.time() # record end time
print("Total:", total)
print(f"Execution Time: {end_time - start_time:.4f} seconds") # .4f → means “show 4 digits after the decimal point
```

```python
#copy() and view
#COPYY

# a copy will create a new independent array in a diff memory location
# meaning if you make any changes in copied_array it won't effect original_array
org_arr = np.array([10,20,30,40,50])
copied_arr = org_arr.copy()
print(org_arr)
print(copied_arr)
```

```python
#copy() and view
# (VIEWW)

# a view is the representation af original_array and it shares same memory location
# meaning if you make any changes in copis_array it will effect in orginal_array
org_arr = np.array([10,20,30,40,50])
copied_arr = org_arr.view()
print(org_arr)
print(copied_arr)
```

```python
# copy() and view()

# copy()
# a copy will create a new independent array in a diff memory location
# meaning if you make any changes in copied_array it won't effect original_array

org_arr = np.array([10,20,30,40,50,60])
print(org_arr)

copied_arr = org_arr.copy()
print(copied_arr)

# let's make some changes in copied_arr
copied_arr[1] = 200
print(copied_arr)
print(org_arr)
```

```python
org_arr = np.array([10,20,30,40,50,60])

print(org_arr)

# view()

# a view is the representation of original_array and it shares same memory location

# meaning if you make any changes in copied_array it will effect original_array

copied_arr = org_arr.view()

print(copied_arr)

# let's make some changes in copied_arr

copied_arr[1] = 200

print(copied_arr)

print(org_arr)
```

```python
org_arr = np.array([10,20,30,40,50,60])
copied_arr = org_arr.view()
# shares_memory() --> return True if arrays are sharing same memory location
print(np.shares_memory(org_arr,copied_arr))
```

```python
arr = np.array([10, 32, 30, 50, 20, 82, 91, 45])
# Find indices where value is 30
indices = np.where(arr == 30)
print(f"Indices where value is 30: {indices}")
# Find indices where values are even
even_indices = np.where(arr % 2 == 0)
print(f"Indices of even values: {even_indices}")
# Conditional selection
new_arr = np.where(arr > 40, arr * 2, arr)
print(f"Array after conditional selection: {new_arr}")
# if array more than 40 needs to multiply by 2 so the answer will be:
# [ 10  32  30 100  20 164 182  90]
```
